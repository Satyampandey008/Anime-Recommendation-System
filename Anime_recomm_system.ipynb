```python
# define the function to split tags into a list of words
def clean_tags(tags_str):
    # convert the tags column values to string (in case they are not already strings)
    tags_str = str(tags_str)
    # remove spaces from the string and split it into a list of words using comma as delimiter
    tags_list = tags_str.replace(' ', '').split(',')
    return tags_list

# apply the function to each row of the 'Tags' column
anime['Tags'] = anime['Tags'].apply(clean_tags)

# print the cleaned tags column
print(anime['Tags'])
```

```python
anime.head()
```

```python
anime['Description'][0]
```

```python
# to convert all the int values inrto string 
anime['Description'] = anime['Description'].astype(str)
```

```python
#converting the 'Description' as a lsit of words 
anime['Description']= anime['Description'].apply(lambda x:x.split())
```

```python
anime.head(50)
```

```python
#to remove whitespaces to avoid confusion of our model 
# ex . sakura haruno = sakuraharuno to avoid confusion with any other character of other anime
anime['Tags'] = anime['Tags'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
anime['Description'] = anime['Description'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
anime.head(50)
```

```python
anime['Labels'] = anime['Tags'] + anime['Description']
```

```python
anime.head()
```

```python
new_df = anime[['Name','Episodes','Rating','Release_year','Labels']]
```

```python
new_df['Labels']=new_df['Labels'].apply(lambda x:" ".join(x))
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
       
    return " ".join(y)
```

```python
new_df['Labels'].apply(stem)
```

```python
new_df['Labels']= new_df['Labels'].apply(stem)
```

```python
new_df['Labels'][0]
```

```python
# converting all alphabets into lower case as it is recommended
new_df['Labels']=new_df['Labels'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['Labels']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(anime):
    anime_index = new_df[new_df['Name'] == anime].index[0]
    distances = similarity[anime_index]
    anime_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
    
    for i in anime_list:
        print(new_df.iloc[i[0]].Name)
        
```

```python
recommend('Fruits Basket the Final Season')
```
